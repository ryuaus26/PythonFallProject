This snippet of code shows the count, mean, standard deviation, min, 25% of the values in the column are less than or equal to this value, 50% which separates the higher and lower half of the data. It is the value below which 50% of the data falls, 75% of the values in the column are less than or equal to this value,and maximum value in the data

```python
#Calcualating summary statistics
data = pd.read_csv('/content/drive/MyDrive/Housing.csv')

summary_stats = data.describe()

print(summary_stats)
```

This snippet of code shows the values of each column in a sorted manner. The intermediate values are not explicitly shown due to the large number of rows in the data

```python
# Sorting the data by columns
columns = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]

for column in columns:
    sorted_data = data.sort_values(column)
    print(f"Sorted data for column {column}:\n{sorted_data}\n")
```

We filter the data through the mean() function. If the value in the data fails to satisfy the mean_value it will be not added. With this we can get rid of data that is lower than the mean_value-

```python
#Filtering data: Only show values that is greater than the mean value in the specified column...

for column_name in data.columns:
    mean_value = data[column_name].mean()
    
    filtered_data = data[data[column_name] > mean_value]
    print(f"Mean value: {mean_value:.3f}")
    print(f"Filtered data for column {column_name}:")
    print(filtered_data)

```

Using the math operation functions in python we can print out the total sum value of the data values in each column and well as its max value and min value. Standard deviation is also accessible through the std() function

```python
#Some math operations on data

for column_name in data.columns:

    
    sum_value = data[column_name].sum()
    
    max_value = data[column_name].max()
    
    min_value = data[column_name].min()
    
    std_deviation = data[column_name].std()
    
    # Print the results for the column
    print(f"Column Name: {column_name}")
    print(f"Sum: {sum_value}")
    print(f"Max: {max_value}")
    print(f"Min: {min_value}")
    print(f"Standard Deviation: {std_deviation}\n")
```

The code plots desired columns. This can be useful when you want to observe the relationships between certain features. Which in this case turns out to be Age and Tax. Through this we can see the correlation between age and tax. 

```python
# Plotting  columns
sns.scatterplot(x='AGE', y='TAX', data=data)
plt.xlabel('AGE')
plt.ylabel('TAX')
plt.title('AGE vs TAX')
plt.show()
```

This code snippet generates a histogram to visualize the distribution of values in the specified column. The x-axis represents the values in the column, while the y-axis represents the frequency or count of those values. The histogram provides insights into the shape and spread of the data distribution for the given column.

```python
#View the distribution of the data with a histogram
column_name = 'CRIM'
plt.hist(data[column_name], bins=10)
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Distribution of {column_name}')
plt.show()
```

this code allows you to compare the mean values of the 'RM' column across different groups defined by the 'CHAS' column. It helps you visualize any differences or patterns in the data.

```python
#Aggregate data
grouped_data = data.groupby('CHAS')['RM'].mean()
grouped_data.plot(kind='bar')
plt.xlabel('Grouping Column')
plt.ylabel('Mean Aggregate Column')
plt.title('Aggregate Data')
plt.show()
```
